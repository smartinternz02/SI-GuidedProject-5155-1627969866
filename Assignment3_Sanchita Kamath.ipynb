```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris=sns.load_dataset("iris")
```

```python
iris
```

## Sepal and Petal

##### Distribution of flower parts

```python
figure= plt.figure(figsize = (12,12))

figure.add_subplot(2,2,1)
plt.title("Sepal Width")
sns.distplot(iris["sepal_width"])

figure.add_subplot(2,2,2)
plt.title("Sepal Length")
sns.distplot(iris["sepal_length"])

figure.add_subplot(2,2,3)
plt.title("Petal Width")
sns.distplot(iris["petal_width"])

figure.add_subplot(2,2,4)
plt.title("Petal Length")
sns.distplot(iris["petal_length"])
```

Inference:   
        
        1. The maximum Sepal Width is observed at 3.0 units 
        2. The maximum Sepal Length is observed between 5 to 6.5 units
        3. The maximum Petal Width is observed at about 1.75 units(not an 
           accurate demarcation) 
        4. The maximum Petal Length is observed at about 5 units
        
While the graph for the sepal width is a perfect bell curve, which means most flowers have 3 units of sepal width which decreases on either side of that particular measure; yet we can see there are few anomalies in the other graphs. 

For the sepal_length graph, the highest value remains stagnant instead of appexing, which means, that there are a larger number of flowers having a large sepal_length. 

For the petal width graph we can observe a dip close to when the petal width reaches 1 unit, and a high number of flowers having close to 0.25 units of length before it, implying that a very few flowers have the petal width of 1 unit, which gradually appexes as petal width nears 1.75 units and further decreases. 

Further, a similar phenomenon can be observed for petal length, where in there are comparatively few flowers growing with petal length 3 units, a larger number of flowers having close to 2 units of petal length, and highest number of flowers having about 5 units of petal length.

##### Sepal Length v/s Petal Length

```python
sns.lineplot(iris["sepal_length"], iris["petal_length"])
plt.title("Sepal Length v/s Petal Length")
```

Inference: It is mostly a trend that as the sepal length increases, the petal length increases. Although, a dip is seen at about 5.3 (not an accurate demarcation) after which there is a sudden/abrupt increase in the length, which is almost consistently growing, without sudden depressions in value

##### Sepal Width v/s Petal Width

```python
sns.lineplot(iris["sepal_width"], iris["petal_width"])
plt.title("Sepal Width v/s Petal Width")
```

Inference: It is mostly a trend that as the sepal width increases, the petal width decreases. This is almost consistent throughout the plotted graph.

##### Sepal Length v/s Sepal Width 

```python
sns.scatterplot(iris["sepal_length"], iris["sepal_width"], data=iris)
plt.title("Sepal Length v/s Sepal Width")
```

Inference: There are two separate clusters of data, which indicates that flowers having larger sepal length and lower sepal width are grouped together, and flowers having higher sepal width and lower sepal length are clumped together. There are also a few flowers that fit in neither of the groups (outliers).

```python
sns.scatterplot(iris["petal_length"], iris["petal_width"], data=iris)
plt.title("Petal Length v/s Petal Width")
```

Inference: There are two separate clusters of data, which are clearer that the one in the above graph. It indicates that flowers having larger petal length and larger petal width are grouped together, and flowers having lower petal width and lower petal length are clumped together. There is also a gap between petal length of 2 to 3 units, wherein no flowers belong. You can also see this dip in the distribution of the petal length graph above.

## Species 

#### Marking outliers

```python
figure= plt.figure(figsize = (12,12))

figure.add_subplot(2,2,1)
plt.title("Sepal Width")
sns.boxplot(x="species", y="sepal_width", data=iris)

figure.add_subplot(2,2,2)
plt.title("Sepal Length")
sns.boxplot(x="species", y="sepal_length", data=iris)

figure.add_subplot(2,2,3)
plt.title("Petal Width")
sns.boxplot(x="species", y="petal_width", data=iris)

figure.add_subplot(2,2,4)
plt.title("Petal Length")
sns.boxplot(x="species", y="petal_length", data=iris)
```

Inference: There are very few outliers in setosa and virginica, but none in versicolor for the sepal width graph. For the sepal length graph, there are no outliers in the setosa and versicolor flowers, but a distant few in the virginica specie. There are outliers in setosa flowers for the Petal Width, but they don't seem to have lower petal width, only larger; whereas none for the other two species. For the Petal length virginica has no outliers, but setosa and versicolor do. Versicolor doesn't have any higher than generated petal length, but does have lower.

#### Finding highest lowest attibutes categorised by species

```python
colors = {"versicolor":"orange", "setosa":"#61adff", "virginica":"#ee8cff"}

figure= plt.figure(figsize = (15,15))

figure.add_subplot(2,2,1)
plt.title("Sepal Width")
plt.bar(iris["species"],iris["sepal_width"], color="#ffd500")
plt.ylabel('Sepal Width')


figure.add_subplot(2,2,2)
plt.title("Sepal Length")
plt.bar(iris["species"],iris["sepal_length"], color="#ff9a42")
plt.ylabel('Sepal Length')

figure.add_subplot(2,2,3)
plt.title("Petal Width")
plt.bar(iris["species"],iris["petal_width"], color="#42d0ff")
plt.ylabel('Petal Width')


figure.add_subplot(2,2,4)
plt.title("Petal Length")
plt.bar(iris["species"],iris["petal_length"], color="#ff428e")
plt.ylabel('Petal Length')
```

I think it would be safe to say, that in the data that has been representated, the setosa flowers have smaller petal size (width and length) but higher sepal width (lower sepal length).

The versicolor flowers seem to lie constantly in the middle region of every plot, hence it may be inferred that these flowers are medium sized, with moderating petal and sepal size.

The Virginica flowers have the comparitively largest petal size. It's sepal size seems to be larger than the versicolor flowers, but the sepal width is smaller than the setosa flowers; although they have a larger sepal length.

#### Comparing physical attributes, categorised by species 

```python
colors = {"versicolor":"orange", "setosa":"#61adff", "virginica":"#ee8cff"}

figure= plt.figure(figsize = (15,15))

figure.add_subplot(2,2,1)
plt.title("Sepal Length v/s Sepal Width")
plt.scatter(iris["sepal_length"], iris["sepal_width"], 
            c=iris["species"].map(colors))
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')


figure.add_subplot(2,2,2)
plt.title("Petal Length v/s Petal Width")
plt.scatter(iris["petal_length"], iris["petal_width"], 
            c=iris["species"].map(colors))
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')


figure.add_subplot(2,2,3)
plt.title("Sepal Length v/s Petal Length")
plt.scatter(iris["sepal_length"], iris["petal_length"], 
            c=iris["species"].map(colors))
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')


figure.add_subplot(2,2,4)
plt.title("Sepal Width v/s Petal Width")
plt.scatter(iris["sepal_width"], iris["petal_width"], 
            c=iris["species"].map(colors))
plt.xlabel('Sepal Width')
plt.ylabel('Petal Width')
```

Inference: 

    1. For Sepal Width v/s Sepal Length it can be inferred that almost all 
    the setosa flowers have a lower sepal length and higher sepal width, and
    the other two species have a larger sepal length and lower petal length.
    
    2. For Petal Width v/s Petal Length it can be inferred that all the 
    setosa flowers have a lower petal length and lower petal width, the
    versicolor flowers have a medium range of petal length and width, whereas
    almost all the virginica flowers have a high petal length and width.
    
    3. For Sepal Length v/s Petal Length it can be inferred that all the 
    setosa flowers lie in the region of low petal and sepal length, all 
    versicolor petals have medium ranging petal and sepal length and the 
    virginica flowers have a high petal and sepal length.
    
    4. For Sepal Width v/s Petal Width it can be inferred that all the 
    setosa flowers lie in the region of high sepal width but low petal width,
    all versicolor petals have medium ranging petal width and lower sepal width
    while the virginica flowers have a high lower width but low sepal width

###### From the above statistics, we can also figure out that the various species of the iris flower are named based on their physical attributes.
